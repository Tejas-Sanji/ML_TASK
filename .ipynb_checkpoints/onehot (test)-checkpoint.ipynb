```python
import pandas as pd
```

```python
dftest=pd.read_csv('cleantest.csv')
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder()
```

```python
dftest['Gender']=le.fit_transform(dftest['Gender']) 
```

```python
dftest['Married']=le.fit_transform(dftest['Married'])
```

```python
dftest['Education']=le.fit_transform(dftest['Education'])
```

```python
dftest['Self_Employed']=le.fit_transform(dftest['Self_Employed'])
```

```python
dftest['Dependents']=le.fit_transform(dftest['Dependents'])
```

```python
dftest['Property_Area']=le.fit_transform(dftest['Property_Area'])
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
 import numpy as np
```

```python
onehot = OneHotEncoder(dtype=np.int, sparse=True)
```

```python
traingen = pd.DataFrame(
    onehot.fit_transform(dftest[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']])\
    .toarray(),
    columns=['Female','Male','Not_Married','Married','No_Dependent','1_Dependent','2_Dependents','3+_Dependents','Graduate','Not_Graduate','Not_Self_Employed','Self_Employed','Rural','Semiurban',
             'Urban'])
```

```python
traingen.head()
```

```python
traingen.tail()
```

```python
traingen['Loan_ID']=dftest.Loan_ID
```

```python
dftest.info()
```

```python
traingen.info()
```

```python
dffinal=pd.merge(traingen,dftest[['Loan_ID','ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']],on='Loan_ID')
```

```python
dffinal.info()
```

```python
dffinal.columns
```

```python
dffinal=dffinal.drop('Loan_ID',axis=1)
```

```python
dffinal.info()
```

```python
dffinal.to_csv('finaltest.csv')
```
