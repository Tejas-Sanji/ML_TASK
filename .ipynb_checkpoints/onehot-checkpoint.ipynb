```python
import pandas as pd
```

```python
dftrain=pd.read_csv('cleantrain.csv')
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder()
```

```python
dftrain['Gender']=le.fit_transform(dftrain['Gender']) 
```

```python
dftrain['Married']=le.fit_transform(dftrain['Married'])
```

```python
dftrain['Education']=le.fit_transform(dftrain['Education'])
```

```python
dftrain['Self_Employed']=le.fit_transform(dftrain['Self_Employed'])
```

```python
dftrain['Dependents']=le.fit_transform(dftrain['Dependents'])
```

```python
dftrain['Property_Area']=le.fit_transform(dftrain['Property_Area'])
```

```python
dftrain['Loan_Status']=le.fit_transform(dftrain['Loan_Status'])
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
 import numpy as np
```

```python
onehot = OneHotEncoder(dtype=np.int, sparse=True)
```

```python
traingen = pd.DataFrame(
    onehot.fit_transform(dftrain[['Gender','Married','Dependents','Education','Self_Employed','Property_Area']])\
    .toarray(),
    columns=['Female','Male','Not_Married','Married','No_Dependent','1_Dependent','2_Dependents','3+_Dependents','Graduate','Not_Graduate','Not_Self_Employed','Self_Employed','Rural','Semiurban',
             'Urban'])
```

```python
traingen.head()
```

```python
traingen['Loan_ID']=dftrain.Loan_ID
```

```python
dffinal=pd.merge(traingen,dftrain[['Loan_ID','ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History','Loan_Status']],on='Loan_ID')
```
